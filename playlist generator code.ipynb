```python
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QLabel, QLineEdit, QPushButton, QMessageBox

class InputWidget(QWidget):
    def __init__(self):
        super().__init__()

        self.init_ui()

    def init_ui(self):
        layout = QVBoxLayout()

        self.label = QLabel('Enter something:')
        self.input_box = QLineEdit()
        self.button = QPushButton('Get Input')
        self.button.clicked.connect(self.get_user_input)

        layout.addWidget(self.label)
        layout.addWidget(self.input_box)
        layout.addWidget(self.button)

        self.setLayout(layout)

    def get_user_input(self):
        user_input = self.input_box.text()
        QMessageBox.information(self, 'User Input', f'User Input: {user_input}')

if __name__ == '__main__':
    app = QApplication([])
    window = InputWidget()
    window.setWindowTitle('Input Widget Example')
    window.show()
    app.exec_()
```
